* Программа для подсчета суммы диапазона чисел

```python
from_ = int(input('Введите первое число'))
to = int(input('Введите второе число'))
def sum_distance(from_, to):
    if from_ > to:
        from_, to = to, from_
    return sum(range(from_, to + 1))
print(f'Сумма всех чисел от {from_} до {to} ={sum_distance(from_, to)}')
```

* Программа для обрезания и повторения строки

```python
text = input('Введите строку для обработки')
offset = int(input('На сколько символов обрезать строку слева?'))
repetitions = int(input('Сколько раз повторить получившуюся строку?'))
def trim_and_repeat(text, offset=0, repetitions=1):
    trimmed_text = text[offset:]
    result = trimmed_text * repetitions
    return result
print(trim_and_repeat(text, offset, repetitions))
```
