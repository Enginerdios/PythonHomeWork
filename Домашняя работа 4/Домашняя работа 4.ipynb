```python
from datetime import datetime

# Список газет с их датами
newspapers = [
    {"name": "The Moscow Times", "date_str": "Wednesday, October 2, 2002"},
    {"name": "The Guardian", "date_str": "Friday, 11.10.13"},
    {"name": "Daily News", "date_str": "Thursday, 18 August 1977"}
]

# Определяем форматы для каждой газеты
date_formats = {
    "The Moscow Times": "%A, %B %d, %Y",
    "The Guardian": "%A, %d.%m.%y",
    "Daily News": "%A, %d %B %Y"
}

# Переводим каждую дату в объект datetime и выводим результат
for paper in newspapers:
    name = paper["name"]
    date_str = paper["date_str"]

    try:
        dt_object = datetime.strptime(date_str, date_formats[name])
        print(f"{name}: {dt_object}")
    except ValueError as e:
        print(f"{name}: Ошибка парсинга - {e}")
```
