* Программа для вывода центральных букв из слова

```python
print('Программа для вывода букв')
word = input('Введите слово')
print('Введенное слово: ' + word)
length = len(word)
if length % 2 == 0:
    middleLetter = word[length // 2 - 1 : length // 2 + 1]
else:
    middleLetter = word[length // 2]
print('Средняя буква: ', middleLetter)
```

* Программа для подбора пар

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
print('Программа для поиска пар')
if len(boys) == len(girls):
    boys.sort()
    girls.sort()
    print('Идеальные пары:')
    for boy, girl in zip(boys, girls):
        print(f'{boy} и {girl}')
else:
    print('Внимание, кто-то может остаться без пары!')
```
